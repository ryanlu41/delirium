```python
def standardize_column_names(df, keep_length_of_stay=False):
    if "Unnamed: 0" in df.columns:
        df.drop(columns=["Unnamed: 0"], inplace=True)
    if "del_start" in df.columns:
        df.drop(columns=["del_start"], inplace=True)
    if "start" in df.columns:
        df.drop(columns=["start"], inplace=True)
    if "end" in df.columns:
        df.rename(columns={"end" :"LOS"}, inplace=True)
    if "index" in df.columns:
        df.drop(columns=["index"], inplace=True)
    if "patientunitstayid" in df.columns:
        df.rename(columns={"patientunitstayid": "PatientStayID"}, inplace=True)
    if "delirium?" in df.columns:
        df.rename(columns={"delirium?" : "Class"}, inplace=True)
    df.replace(to_replace=[True, False], value=[1, 0], inplace=True)
    if keep_length_of_stay is False:
        df.drop(columns=["Class", "LOS"], inplace=True)
    return df
```
